## Simple Neural Network in PyTorch

```python
# A simple neural network with 1 input and 1 output (no hidden layer)
net = nn.Linear(in_features= 1, out_features= 1)  # -> univariate linear regression model
net
```

```python
input_tensor  = torch.tensor([[0.3]])
input_tensor
```

```python
y = net(input_tensor)
y
```

### Stacking Layers



```python
n_features = 8

n_target = 1

model = nn.Sequential(
            nn.Linear(n_features, 20),
            nn.Linear(20, 15),
            nn.Linear(15, 10),
            nn.Linear(10, 4),
            nn.Linear(4, n_target)
)

model
```

## Optimizers in PyTorch

#### Stochastic Gradient Descent (SGD)

```python
learning_rate = 1e-2 # 0.01 # Typical values range between 1e-2 to 1e-4
optimizer_sgd = torch.optim.SGD(net.parameters(), lr = learning_rate)
```

#### Adaptive Gradient (Adagrad)

```python
optimizer_ada = torch.optim.Adagrad(net.parameters(), lr = learning_rate)
```

#### Root Mean Square Propagation (RMSprop)

```python
optimizer_rms = torch.optim.RMSprop(net.parameters(), lr = learning_rate)
```

#### Adaptive Moment Estimation (Adam)

```python
optimizer_adam = torch.optim.Adam(net.parameters(), lr = learning_rate)
```

See PyTorch documentation for more algorithms: https://pytorch.org/docs/stable/optim.html
